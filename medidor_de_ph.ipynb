Medidor de Ph

```python
ph = int(input("dime el ph:"))

if ph == 7:
  print("es neutro el ph")
elif ph < 7 and ph >= 4:
  print("es poco acido el ph")
elif ph < 4 and ph >= 0:
  print("es muy acido el ph")
elif ph > 7 and ph <= 14:
  print("es poco alcalino el ph")
else:
  print("es muy alcalino el ph")
```
